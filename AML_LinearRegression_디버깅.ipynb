#디버깅 전의 키 175cm인 사람의 몸무게 예측 결과를 보면 사람의 몸무게 값이 아님.

```python
import numpy as np

# 입력 데이터 (키(cm), 몸무게(kg))
X = np.array([170, 175, 180, 165, 160, 185])
y = np.array([68, 75, 80, 55, 50, 90])

# 모델 파라미터 초기화 (w: 가중치, b: 절편)
w = 0.0
b = 0.0

# 학습률(learning rate) 설정
lr = 0.0001

# 예측 함수 정의
def predict(x):
    y_pred = w*x + b
    return y_pred

# 손실 함수 정의 (평균 제곱 오차)
def mse_loss(y_true, y_pred):
    mse = np.mean((y_true - y_pred)**2)
    return mse

# 경사 하강법(Gradient Descent)을 사용하여 손실 함수 최소화
def train(X, y, w, b, lr, epochs):
    for epoch in range(epochs):
        # 예측 값 계산
        y_pred = predict(X)

        # 손실 함수 계산
        loss = mse_loss(y, y_pred)

        # 가중치(w)와 절편(b) 업데이트
        grad_w = np.mean((y_pred - y)*X)
        grad_b = np.mean(y_pred - y)
        w = w - lr*grad_w
        b = b - lr*grad_b

        # 로그 출력
        if epoch % 100 == 0:
            print("Epoch %d: loss=%.4f, w=%.4f, b=%.4f" % (epoch, loss, w, b))

    return w, b

# 학습 실행
w, b = train(X, y, w, b, lr, epochs=1000)
```

```python
# 키가 175cm인 경우 몸무게 예측
my_height = 175
my_weight = predict(my_height)

print("나의 키는 %dcm이고, 예측 몸무게는 %.1fkg입니다." % (my_height, my_weight))
```

```python
# 키가 175cm인 경우 몸무게 예측
my_height = 175
my_weight = predict(my_height, w, b)

print("나의 키는 %dcm이고, 예측 몸무게는 %.1fkg입니다." % (my_height, my_weight))
```

#최초 설정한 predict함수의 인자값이 x만 할당되어 있었기 때문에 w와 b가 업데이트 되지 않았고, 그 결과 loss값에 변동이 없었던 것임.
#이에 predict함수에 w와 b를 추가로 인자로 설정해줌.
#하지만 이번엔 loss값이 갑자기 inf(무한)으로 발산해버림.
#즉, 경사하강법을 사용해 기울기가 0인 지점을 찾아가다가 학습률이 너무 커서 갑자기 반대쪽으로 넘어가서 무한대로 발산한 것임
#학습률 조정으로 해결

```python
import numpy as np

# 입력 데이터 (키(cm), 몸무게(kg))
X = np.array([170, 175, 180, 165, 160, 185])
y = np.array([68, 75, 80, 55, 50, 90])

# 모델 파라미터 초기화 (w: 가중치, b: 절편)
w = 0.0
b = 0.0

# 학습률(learning rate) 설정
lr = 0.000001                         #수정 전 lr = 0.0001

# 예측 함수 정의
def predict(x, w, b):                #수정 전 def predict(x):
    y_pred = w*x + b
    return y_pred

# 손실 함수 정의 (평균 제곱 오차)
def mse_loss(y_true, y_pred):
    mse = np.mean((y_true - y_pred)**2)
    return mse

# 경사 하강법(Gradient Descent)을 사용하여 손실 함수 최소화
def train(X, y, w, b, lr, epochs):
    for epoch in range(epochs):
        # 예측 값 계산
        y_pred = predict(X, w, b)         #수정 전 def predict(x):

        # 손실 함수 계산
        loss = mse_loss(y, y_pred)

        # 가중치(w)와 절편(b) 업데이트
        grad_w = np.mean((y_pred - y)*X)
        grad_b = np.mean(y_pred - y)
        w = w - lr*grad_w
        b = b - lr*grad_b

        # 로그 출력
        if epoch % 100 == 0:
            print("Epoch %d: loss=%.4f, w=%.4f, b=%.4f" % (epoch, loss, w, b))

    return w, b

# 학습 실행
w, b = train(X, y, w, b, lr, epochs=1000)
```

```python
# 키가 175cm인 경우 몸무게 예측
my_height = 175
my_weight = predict(my_height, w, b)

print("나의 키는 %dcm이고, 예측 몸무게는 %.1fkg입니다." % (my_height, my_weight))
```
